```python
import numpy as np
import matplotlib.pyplot as plt
from control import tf, feedback, step_response, pzmap

# Define the open-loop transfer function G(s)
# G(s) = 1 / [(s+1)(s+2)(s+3)]
numerator = [1]
denominator = np.polymul([1, 1], np.polymul([1, 2], [1, 3]))
G = tf(numerator, denominator)

# Define the proportional controller C(s) = Kp
# We will vary Kp later in the loop

# Set up the figure for plotting
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
step_plot, root_locus_plot = axs

# Step response for different values of Kp
Kp_values = [0.1, 1, 5, 10]  # Gains to visualize
time = np.linspace(0, 10, 500)  # Time vector for step response

for Kp in Kp_values:
    # Closed-loop transfer function: T(s) = Kp * G(s) / (1 + Kp * G(s))
    C = tf([Kp], [1])  # Controller
    T = feedback(C * G, 1)  # Closed-loop transfer function
    
    # Simulate the step response
    t, y = step_response(T, time)
    step_plot.plot(t, y, label=f'Kp={Kp}')
    
    # Plot the poles on the root locus
    poles = np.roots(np.polyadd(denominator, [Kp]))
    root_locus_plot.scatter(poles.real, poles.imag, label=f'Kp={Kp}')

# Formatting for step response plot
step_plot.set_title('Step Response')
step_plot.set_xlabel('Time (s)')
step_plot.set_ylabel('Response')
step_plot.legend()
step_plot.grid()

# Formatting for root locus/pole plot
root_locus_plot.set_title('Poles in the Real-Imaginary Plane')
root_locus_plot.set_xlabel('Real')
root_locus_plot.set_ylabel('Imaginary')
root_locus_plot.axhline(0, color='black', lw=0.5, linestyle='--')  # Imaginary axis
root_locus_plot.axvline(0, color='black', lw=0.5, linestyle='--')  # Real axis
root_locus_plot.legend()
root_locus_plot.grid()

# Show the plots
plt.tight_layout()
plt.show()
```
